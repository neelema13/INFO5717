```python
import csv
import statistics
#from numpy import median
from scipy import stats

def readfilecontents(file_name):
    CLIMATE_STATS = []
    print("Let's open this file using a dictionary")
    with open(file_name, 'r', encoding='utf8', newline='') as f:
        reader = csv.DictReader(f, delimiter=',')
                #no need to skip the header row when using a dictionary
        for row in reader:
            year = row["Day"]
            field_name = int(row["High"])
            value = int(row["Low"])
            precip = int(row["Precip"])
            denton_dict = {"Day": year, "High": field_name, "Low": value, "Precip":precip}
            print("the values for each of the rows is----------", year, field_name, value, precip)
            CLIMATE_STATS.append(denton_dict)
    return CLIMATE_STATS
```

```python
def t_test_mean(housing_stats, field):
    t_test_mean_list = []
    for item in housing_stats:
        t_test_mean_list.append(int(item[field]))
    m = statistics.mean(t_test_mean_list)
    return m
def t_test_median(housing_stats, field):
    t_test_mean_list = []
    for item in housing_stats:
        t_test_mean_list.append(int(item[field]))
    med = statistics.median(t_test_mean_list)
    return med

def t_test_stddev(housing_stats, field):
    t_test_mean_list = []
    for item in housing_stats:
        t_test_mean_list.append(int(item[field]))
    s = statistics.stdev(t_test_mean_list)
    return s
```

```python
def get_t_test_low(housing_stats):
    t_test_low = []
    for i in housing_stats:
        t_test_low.append(i["Low"])
    return t_test_low

def get_t_test_high(housing_stats):
    t_test_high = []
    for i in housing_stats:
        t_test_high.append(i["High"])
    return t_test_high

def get_t_test_precip(housing_stats):
    t_test_precip = []
    for i in housing_stats:
        t_test_precip.append(i["Precip"])
    return t_test_precip
```

```python
CLIMATE_VAL = readfilecontents('/Users/neelemaparwani/2011-Jan-June.csv')
R_LOW1 = get_t_test_low(CLIMATE_VAL)
print("the rvs for the low temp is-----------", R_LOW1, len(R_LOW1))
R_HIGH1 = get_t_test_high(CLIMATE_VAL)
print("the rvs for the high temp sample1 is----------", R_HIGH1)
R_PRECIP1 = get_t_test_precip(CLIMATE_VAL)
print("the rvs for the precip temp sample1 is----------", R_PRECIP1)
LOW_MEAN = t_test_mean(CLIMATE_VAL, "Low")
LOW_STDEV = t_test_stddev(CLIMATE_VAL, "Low")
HIGH_MEAN = t_test_mean(CLIMATE_VAL, "High")
HIGH_STDEV = t_test_stddev(CLIMATE_VAL, "High")
PRECIP_MEAN = t_test_mean(CLIMATE_VAL, "Precip")
PRECIP_STDEV = t_test_stddev(CLIMATE_VAL, "Precip")
PRECIP_MEDIAN = t_test_median(CLIMATE_VAL, "Precip")
HIGH_MEDIAN = t_test_median(CLIMATE_VAL, "High")
LOW_MEDIAN = t_test_median(CLIMATE_VAL, "Low")
```
