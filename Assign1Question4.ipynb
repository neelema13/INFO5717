```python

#Taking 2 sentences as input

str1=input("Enter first sentence ")
str2=input("Enter second sentence ")
print("\n")


print("The first sentence is",str1)
print("The second sentence is",str2)


#Converting a string into a list
print("\n")


#split() method is found at https://stackoverflow.com/questions/8266529/python-convert-string-to-list
list1=str1.split()
list2=str2.split()
print("The first list is", list1)
print("The second list is", list2)


#Function for removing the punctuation marks from the given string
print("\n")

def string_without_punctuation(oldstring):

	punctuationStr="!,.*&:;"
	newStr=" "
	for s in oldstring:
		if s not in punctuationStr:
			newStr=newStr+s
	return newStr
	

newStr1=" "
newStr1=string_without_punctuation(str1)
print("Sentence 1 without punctuation is :", newStr1)

newStr2=" "
newStr2= string_without_punctuation(str2)
print("Sentence 2 without punctuation is :", newStr2)


#This is the code for Quest5.3
print("\n")

print("The words that exist in both the strings are :")

#split() method is found at https://stackoverflow.com/questions/8266529/python-convert-string-to-list
c=0
list3=newStr1.split()
list4= newStr2.split()
for w in list3:
	for w1 in list4:
		if w.lower()==w1.lower():
			c=c+1
			print(w)
if c==0:
	print("No match found")



#This is the code for Quest 5.5
print("\n")


#function defined for calculating the number of vowels
def calculate_vowels(userString):
	c1=0
	c2=0
	c3=0
	c4=0
	c5=0
	
	for s in userString:
		if s=='a' or s=='A':
			c1=c1+1
		elif s=='e' or s=='E':
			c2=c2+1
		elif s=='i' or s=='I':
			c3=c3+1
		elif s=='o' or s=='O':
			c4=c4+1
		elif s=='u' or s=='U':
			c5=c5+1

	print('For sentence "',userString,'" the vowels are')
	print("a:",c1)
	print("e:",c2)
	print("i:",c3)
	print("o:",c4)
	print("u:",c5)


#calling the function for both the strings to calculate the number of vowels
calculate_vowels(str1)
calculate_vowels(str2)
```
